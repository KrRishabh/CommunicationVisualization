```python
import pandas as pd
import json

timelineData = open('timelineData.json')
data = json.load(timelineData)
```

```python
data_frame = pd.DataFrame(columns=['timestamp', 'user_id', 'zoom_id', 'email', 'name'])
timelines = data["timeline"]

for timeline in timelines:

    timestamp = timeline["ts"]
    users = timeline["users"]

    if len(users):
        user_id = users[0]["user_id"]
        zoom_id = users[0]["zoom_userid"]
        email = users[0]["email_address"]
        name = users[0]["username"]
        
        data_frame = data_frame.append({"timestamp": timestamp , "user_id":user_id , 'zoom_id':zoom_id, 'email':email,
                                'name':name},ignore_index = True)
```

```python
def getSeconds(time):
  x = time.split(":")
  res = int(x[0])*60*60 + int(x[1])*60 + float(x[2])
  return res

def getDuration(startTime, endTime):
  return ( getSeconds(endTime) - getSeconds(startTime) )

    



prevUser = data['timeline'][0]['users'][0]['username']
prevTimestamp = data['timeline'][0]['ts']
result = []
for datapoint in data['timeline']:
  if( len(datapoint['users']) != 0 and datapoint['users'][0]['username'] != prevUser) :
    # timediff= prevTimestamp +" to "+ datapoint['ts']
    timediff = getDuration(prevTimestamp,datapoint['ts'])
    obj = {"from":prevUser, "to":datapoint['users'][0]['username'], 'duration':timediff }
    result.append(obj)
    prevUser = datapoint['users'][0]['username']
    prevTimestamp = datapoint['ts']

print(result)
```
